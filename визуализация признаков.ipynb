Задание: Используя датасет, визуализировать различными способами минимум три признака, а также минимум три взаимодействия между признаками. Результат можно прислать в виде ноутбука/ссылки на ноутбук, а также в виде готовых графиков.

Name - Название игры

Platform - Платформа игры (PC,PS4 и тд.)

Year - Год релиза

Genre - Жанр игры

Publisher - Издатель

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

Critic_score - Aggregate score compiled by Metacritic staff

Criticcount - The number of critics used in coming up with the Criticscore

User_score - Score by Metacritic's subscribers

Usercount - Number of users who gave the userscore

Developer - Party responsible for creating the game

Rating - Возрастной рейтинг

```python
df.info()
```

```python
df['User_Score'] = pd.to_numeric(df['User_Score'],errors='coerce')
df['User_Score'].dtypes
```

```python
df.describe()
```

```python
# Визуализация разбивки игр по жанрам
plt.figure(figsize=(15,4))
sns.countplot(x='Genre', data=df, order=df['Genre'].value_counts().index)
plt.show()
```

```python
values = df['Genre'].value_counts().values
labels = df['Genre'].value_counts().index
plt.figure(figsize=(10,10))
plt.pie(values, labels=labels)
plt.show()
```

```python
# Визуализация разбивки игр по платформам
plt.figure(figsize=(15,4))
sns.countplot(x='Platform', data=df, order=df['Platform'].value_counts().index)
plt.show()
```

```python
# Визуализация выпусков игр в различные года
df['Year_of_Release'].hist()
plt.xlabel('Year')
plt.ylabel('count')
plt.show()
```

```python
# Визуализация оценок Critic score & User score 
plt.figure(figsize=(15,4))
plt.subplot(1, 2, 1)
sns.boxplot(x='Critic_Score', data=df)
plt.subplot(1, 2, 2)
sns.boxplot(x='User_Score', data=df)
plt.show()
```

```python
# Рассмотрим, различаются ли по возрастному рейтингу игры разных жанров
plt.figure(figsize=(15,10))
sns.countplot(x='Genre',hue='Rating',data = df)
plt.show()
```

```python
# Проверим скоррелированность признаков
sns.heatmap(df.corr())
plt.show()
```

```python
# Рассмотрим, различаются ли оценки критиков и пользователей
plt.scatter(df['Critic_Score'],df['User_Score'], color='orange',alpha=0.2)
plt.show()
```

```python
# Изобразим диаграмму рассеивания для уточнения справедливости оценок
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].scatter( df['Critic_Count'], df['Critic_Score'])
axes[1].scatter(df['User_Count'], df['User_Score'], )
axes[0].set_xlabel('Critic count')
axes[1].set_xlabel('User count')
axes[0].set_ylabel('Critic score')
axes[1].set_ylabel('User score')
print('Распределение оценок в зависимости от количества оценивающих')
plt.show()
```

```python
# Рассмотрим, различаются ли оценки пользователей для разных жанров, платформ и рейтингов
plt.figure(figsize=(20,7))
sns.boxplot(x='Genre',y='User_Score',data=df)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.boxplot(x='Platform',y='Critic_Score',data=df)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.violinplot(x='Rating',y='User_Score',data=df)
plt.show()
```

```python
# Рассмотрим, как связаны продажи с годом релиза
fig = plt.figure(figsize=(15,10))
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.15, 0.56, 0.4, 0.3])
axes1.scatter(df['Year_of_Release'], df['Global_Sales'])
axes1.set_xlabel('Year')
axes1.set_ylabel('Global Sales')
axes2.hist(df['Year_of_Release'],color='red', alpha=0.2)
axes2.set_xlabel('Year')
axes2.set_ylabel('count')
plt.show()
```

```python
# Рассмотрим, как связаны продажи с оценками
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].scatter(df['Critic_Score'], df['Global_Sales'])
axes[1].scatter(df['User_Score'], df['Global_Sales'])
axes[0].set_xlabel('Critic score')
axes[1].set_xlabel('User score')
axes[0].set_ylabel('Global Sales')
axes[1].set_ylabel('Global Sales')
plt.show()
```
